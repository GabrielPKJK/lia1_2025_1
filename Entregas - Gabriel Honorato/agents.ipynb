```python
from dotenv import load_dotenv
import os
import google.generativeai as genai
```

```python
# Carregar vari√°vel de ambiente
load_dotenv()
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
for model in genai.list_models():
    print(model.name)
```

```python
# Fun√ß√£o para criar agentes simulando comportamentos
def gerar_resposta_gemini(prompt):
    model = genai.GenerativeModel("models/gemini-1.5-flash")
    response = model.generate_content(prompt)
    return response.text.strip()
```

```python
# Agente: Planejador de Dieta
def planejar_dieta(objetivo):
    prompt = f"""
    Voc√™ √© um nutricionista especializado. Com base no objetivo "{objetivo}", crie um plano alimentar di√°rio completo, incluindo:
    - Caf√© da manh√£
    - Almo√ßo
    - Lanche(s)
    - Jantar
    O plano deve ser balanceado, adequado ao objetivo, e conter sugest√µes realistas.
    """
    return gerar_resposta_gemini(prompt)
```

```python
# Agente: Calculadora de Macros
def calcular_macros(dieta):
    prompt = f"""
    Com base na dieta abaixo, calcule uma estimativa nutricional com:
    - Calorias totais
    - Quantidade aproximada de prote√≠nas, carboidratos e gorduras (em gramas)
    
    DIETA:
    {dieta}
    """
    return gerar_resposta_gemini(prompt)
```

```python
objetivo = "ganhar massa muscular"
dieta = planejar_dieta(objetivo)
macros = calcular_macros(dieta)
```

```python
# Resultados
print("üçΩÔ∏è Plano Alimentar:\n")
print(dieta)

print("\nüìä Estimativa Nutricional:\n")
print(macros)
```
